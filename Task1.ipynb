```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
```

```python
def etta1(n, array)->float:
    return 2*sum(array[:n])/n

def etta2(n, array)->float:
    return (n+1)*min(array[:n])

def etta3(n, array)->float:
    return (n+1)/n*max(array[:n])

def etta4(n, array)->float:
    return max(array[:n]) + min(array[:n])

def etta5(n, array):
    return array[0] + 1/(n-1)*sum(array[1:n])
```

```python
N = 500
O = 200
values = sps.uniform.rvs(size=N)*O
```

```python
etta_1 = abs(np.array([etta1(i, values) for i in range(1, len(values)+1)]) - O)
etta_2 = abs(np.array([etta2(i, values) for i in range(1, len(values)+1)]) - O)
etta_3 = abs(np.array([etta3(i, values) for i in range(1, len(values)+1)]) - O)
etta_4 = abs(np.array([etta4(i, values) for i in range(1, len(values)+1)]) - O)
etta_5 = abs(np.array([etta5(i, values) for i in range(2, len(values)+2)]) - O)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter([0],[0], color = "white", label = f'etta = {O}')
ax.plot(np.arange(1, N+1), etta_1, color = 'orange', label= '1) 2*(sum of X_n)/n')
#ax.plot(np.arange(1, N+1), etta_2, color = 'green', label = '(n+1)*(min of X_n)')
ax.plot(np.arange(1, N+1), etta_3, color = 'red', label = '3) (n+1)/n*(max of X_n)')
ax.plot(np.arange(1, N+1), etta_4, color = 'blue', label = '4) (max of X_n) + (min of X_n)')
ax.plot(np.arange(1, N+1), etta_5, color = 'green', label = '5) X_1 + 1/(n-1)*(sum from X_2 to X_n)')


ax.set_title('T1')
ax.set_ylabel('error')
ax.set_xlabel('count of samples')
ax.legend(loc='upper right', shadow=True, framealpha=1, edgecolor='black')
plt.savefig('saved_figure.png')
 
plt.show()
```
